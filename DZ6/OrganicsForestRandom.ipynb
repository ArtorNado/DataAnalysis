```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score, train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('C:/Users/One/Downloads/Dataset/dataframe_ORGANICS/ORGANICS.csv', delimiter=',')
data.head()
```

```python
print(data.shape) 
print(data.info())
```

```python
data['TargetBuy'].replace([0,1],['No','Buy'],inplace=True)
data
```

```python
data1= data.drop(data.columns[[0,12]], axis=1)
```

```python
# for each column, get value counts in decreasing order and take the index (value) of most common class
data4 = data1.apply(lambda x: x.fillna(x.value_counts().index[0]))
print(data4.info())
```

```python

#import the necessary module
from sklearn import preprocessing
# create the Labelencoder object
le = preprocessing.LabelEncoder()
#convert the categorical columns into numeric
data4['DemClusterGroup'] = le.fit_transform(data4['DemClusterGroup'])
data4['DemGender'] = le.fit_transform(data4['DemGender'])
data4['DemReg'] = le.fit_transform(data4['DemTVReg'])
data4['DemTVReg'] = le.fit_transform(data4['DemTVReg'])
data4['PromClass'] = le.fit_transform(data4['PromClass'])
data4['TargetBuy'] = le.fit_transform(data4['TargetBuy'])
#display the initial records
data4.head(20)
```

```python
x=data4.values[:, 0:10]
y=data4.values[::,10]
print(x)
print(y)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.30, random_state = 10)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble  import  RandomForestClassifier 
tree_acc = pd.DataFrame(columns=['tree', 'score', 'Misclassified samples'])
n = 1
while n < 302 :
    rf = RandomForestClassifier(n_estimators=n,  max_features = 'sqrt',max_depth=6, min_samples_split=2, random_state=42)
    # Обучаем на тренировочных данных
    rf.fit(X_train, y_train)
    # Действующая классификация
    y_pred = rf.predict(X_test)
    n += 10
    count_misclassified = (y_test != y_pred).sum()
    tree_acc = tree_acc.append({'tree': n, 'score': rf.score(X_test, y_test),
                                'Misclassified samples': format(count_misclassified)}, ignore_index=True)
```

С 100 дерева точность практически не меняется, следует 100 - опитмальное кол-во деревьев

```python
max_depth_grid = [i for i in range(3, 30, 3)]

tree_acc  = pd.DataFrame(columns=['depth', 'score'])
for max_depth in max_depth_grid:
    rfc = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, oob_score=True, max_depth=max_depth)
    rfc.fit(X_train, y_train)
    rfc.score(X_test, y_test)
    tree_acc  = tree_acc .append({'depth': max_depth, 'score': rfc.score(X_test, y_test)}, ignore_index=True)
    
plt.plot(tree_acc ['depth'], tree_acc ['score'])
```

Максимальная точность при значении глубины = 10, далее точноть понижается 

```python
min_samples_split_grid = [i for i in range(2, 10, 1)]

tree_acc = pd.DataFrame(columns=['samples_split', 'score'])
for min_samples_split in min_samples_split_grid:
    rfc = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, oob_score=True, min_samples_split=min_samples_split)
    rfc.fit(X_train, y_train)
    rfc.score(X_test, y_test)
    tree_acc = tree_acc.append({'samples_split': min_samples_split, 'score': rfc.score(X_test, y_test)}, ignore_index=True)
    
plt.plot(tree_acc['samples_split'], tree_acc['score'])
```

Наибольшая точность достигается при кол-ве разбиений равным 9
