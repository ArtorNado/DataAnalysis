```python
more10 = data[data['absences'] > 10]['G3'].mean()
less10 = data[data['absences'] <= 10]['G3'].mean()
if(more10 > less10): print('Нет, не правда')
else: print('Да, правда') 
```

```python
# Кол-во часов учебы в неделю
study1 = data[(data['studytime'] == 1) | (data['studytime'] == 2) ]['G3'].mean()
study2 = data[(data['studytime'] == 3) | (data['studytime'] == 4) ]['G3'].mean()
if(study1 > study2): print('Люди, которые тратят на обучение до 5 часов в неделю, в среднем учатся лучше, чем те, кто учится больше 5 ')
else: print('Люди, которые тратят на обучение больше 5 часов в неделю, в среднем учатся лучше, чем те, кто учится меньше 5')
    
# Доп занятия    
schoolsup1 = data[data['schoolsup'] == 'yes']['G3'].mean()
schoolsup2 = data[data['schoolsup'] == 'no']['G3'].mean()
if(schoolsup1 > schoolsup2): print('Люди, которые посещают дополнительные занятия, в среднем учатся лучше')
else: print('Люди, которые не посещают дополнительные занятия, в среднем учатся лучше')
    
# Доп занятия платные    
paid1 = data[data['paid'] == 'yes']['G3'].mean()
paid2 = data[data['paid'] == 'no']['G3'].mean()
if(paid1 > paid2): print('Люди, которые посещают дополнительные платные занятия, в среднем учатся лучше')
else: print('Люди, которые не посещают дополнительные платные занятия, в среднем учатся лучше')
    
# Внеклассная деятельность    
activities1 = data[data['activities'] == 'yes']['G3'].mean()
activities2 = data[data['activities'] == 'no']['G3'].mean()
if(activities1 > activities2): print('Люди, которые занимаются внеклассной деятельностью, в среднем учатся лучше')
else: print('Люди, которые не занимаются внеклассной деятельностью, в среднем учатся лучше')
    
# Кол-во свободного времени после школы
freetime1 = data[(data['freetime'] == 1) | (data['freetime'] == 2) ]['G3'].mean()
freetime2 = data[(data['freetime'] == 3) | (data['freetime'] == 4) | (data['freetime'] == 5) ]['G3'].mean()
if(freetime1 > freetime2): print('Люди, у которых мало свободного времени после школы, в среднем учатся лучше')
else: print('Люди, у которых средне или много свободного времени после школы, в среднем учатся лучше')
    
# Романтические отношения 
romantic1 = data[data['romantic'] == 'yes']['G3'].mean()
romantic2 = data[data['romantic'] == 'no']['G3'].mean()
if(romantic1 > romantic2): print('Люди, которые cсостоят в отношениях, в среднем учатся лучше')
else: print('Люди, которые не состоят в отношениях, в среднем учатся лучше')
```
